```python
# import all necessary libraries
import cv2
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image
```

```python
def show_multi_img(images , titles , row , column , vmin=0 , vmax=255):
    fig = plt.figure(figsize=(15,17) , dpi=100)

    for i in range(len(images)):
        fig.add_subplot(row , column , i+1)
        plt.imshow(images[i] , cmap='gray' , vmin=vmin , vmax=vmax)
        plt.axis('off')
        plt.title(titles[i])
```

```python
#read image
img = cv2.imread(r'sample.jpg' , 0)
img_pil = Image.open(r'sample.jpg')
```

```python
show_multi_img([img , img_pil] , ['img' , 'img_pil'] , 1 , 2)
```

```python
#function that plot histogram of image
def display_hist_plt (img , bins=256 , range=(0,256)):
    plt.figure(figsize=(4,2) , dpi=100)
    plt.hist(img.flat , bins=bins , range=range , color='red')
    plt.show()
```

```python
display_hist_plt(img)
```

###  stretching

<img src="stretching1.png">

<img src="stretching2.png">

```python
# np.uint8(((img-min_old)/(max_old-min_old))*255)

min_new = 0
max_new = 255 
min_old = np.min(img)
max_old = np.max(img)

cs_img = np.uint8(((img-min_old)/(max_old-min_old))*255)
```

```python
show_multi_img([img , cs_img] , ['img' , 'cs_img'] , 1 , 2)
```

```python
display_hist_plt(cs_img)
```

###  Histogram Equalization

```python
hi_eq = cv2.equalizeHist(img)
```

```python
show_multi_img([img , hi_eq] , ['img' , 'hi_eq'] , 1 , 2)
```

```python
display_hist_plt(hi_eq)
```

```python
cv2.imwrite('output.png' , hi_eq)
```

###  Gamma

```python
# np.uint8(((img/255)**gamma)*255)
gamma = 0.1
gamma_img = np.uint8(((img/255)**gamma)*255)
```

```python
show_multi_img([img , gamma_img] , ['img' , 'gamma_img'] , 1 , 2)
```

```python
display_hist_plt(gamma_img)
```

###  Histogram Equalization & Gamma

```python
gamma = 0.3
he_gamma_img = np.uint8(((hi_eq/255)**gamma)*255)
```

```python
show_multi_img([img , he_gamma_img] , ['img' , 'he_gamma_img'] , 1 , 2)
```

```python
display_hist_plt(he_gamma_img)
```
