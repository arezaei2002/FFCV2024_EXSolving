```python
# import all necessary libraries
import numpy as np 
import matplotlib.pyplot as plt
import cv2 
import os
```

<p>1. Read all content of the folder (/dataset/images), Regardless of the file type and file name in the folder using python. (use os.scandir() or os.listdir() )[40] </p>

```python
path = '/Users/arezaie/Desktop/FanFil_CV_Dec2023/Ex_solving/Ex5/images'
```

```python
os.listdir(path)
```

```python
fileList = [] # store file name

for file in os.listdir(path):
    fileList.append(os.path.join(path,file))
```

```python
fileList
```

<p>2. Check the type of image ( Color, gray or binary).[70] </p>

```python
def check_image_type(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

    # Check for invalid image
    if image is None:
        return "Invalid Image"

    # Check if the image is color
    if len(image.shape) == 3:
        return "Color Image"

    # Check if the image is binary
    if len(np.unique(image)) == 2:
        return "Binary Image"

    # If it's not color or binary, then it's grayscale
    return "Grayscale Image"
```

```python
for file in fileList:
    image_type = check_image_type(file)
    print(f"The image is a {image_type}")
    
```

<p>3. If the image is in color, convert it to grayscale, and then find the opposite of the image
( 255 - pixel intensity). [70] </p>

```python
def inverted_img(img):
    inverted_image = 255 - img
    return inverted_image
```

```python
for file in fileList:
    if check_image_type(file) == 'Color Image':
        img = cv2.imread(file)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        plt.subplot(1, 2, 1)
        plt.imshow(gray,cmap='gray')
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        inverted_gray = inverted_img(gray)
        plt.imshow(inverted_gray,cmap='gray')
        plt.title('Inverted Image')
        plt.axis('off')

        plt.show()  
```

<p>4. If the image is in gray, convert it to binary, and then find the opposite of the image (255 - pixel intensity).[70] </p>

```python
for file in fileList:
    if check_image_type(file) == 'Grayscale Image':
        img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

        # Convert grayscale image to binary
        _, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)


        plt.subplot(1, 2, 1)
        plt.imshow(img , cmap='gray')
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        inverted_binary = inverted_img(binary_img)
        plt.imshow(inverted_binary  , cmap='gray')
        plt.title('Inverted Image')
        plt.axis('off')

        plt.show()  
```

<p>5. If the image is binary, find the opposite of the image (255 - pixel intensity).[70] </p>

```python
for file in fileList:
    if check_image_type(file) == 'Binary Image':
        img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)

        plt.subplot(1, 2, 1)
        plt.imshow(img, cmap='gray')
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        inverted_binary = inverted_img(img)
        plt.imshow(inverted_binary, cmap='gray')
        plt.title('Inverted Image')
        plt.axis('off')

        plt.show()
```
