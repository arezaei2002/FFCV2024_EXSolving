## Import necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_multi_img (rows,columns,img_names,titles = None,vmin = 0,vmax = 255):
    fig = plt.figure(figsize=(15,17),dpi=100)
    for i in range(len(img_names)):
        fig.add_subplot(rows,columns,i+1)
        plt.imshow(img_names[i],cmap='gray',vmin=vmin,vmax=vmax)
        plt.axis('off')
        if titles is not None:
            plt.title(str(i+1) + '. ' + titles[i])
    plt.show()
```

```python
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6, vmin=0, vmax=255):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map, vmin=vmin, vmax=vmax)
```

1. Find and plot the Circles in the image (b.jpg), for pixels with intensity greater than 120

```python
img = cv2.imread('media/b.jpg',-1)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show_img_plt(gray_img)
```

```python
gauss_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
```

```python
ret, thresh = cv2.threshold(gauss_img, 120, 255, cv2.THRESH_BINARY_INV)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
contour_img = img.copy()
contour_img = cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB)
```

```python
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 10);
```

```python
show_img_plt(contour_img)
```

2. Find the shape of the following object using contours. (400)

```python
img = cv2.imread('media/F6.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show_img_plt(gray_img)
```

```python
gauss_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
```

```python
ret, thresh = cv2.threshold(gauss_img, 165, 255, cv2.THRESH_BINARY_INV)
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
contour_img = img.copy()
contour_img = cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB)
```

```python
cv2.drawContours(contour_img, contours, 2, (225, 255, 0), 2);
cv2.drawContours(contour_img, contours, 2, (0, 0, 0), 1);
cv2.drawContours(contour_img, contours, 3, (225, 255, 0), 2);
cv2.drawContours(contour_img, contours, 3, (0, 0, 0), 1);
```

```python
show_img_plt(contour_img)
```
